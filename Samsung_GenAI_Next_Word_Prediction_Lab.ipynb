```
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch
import gradio as gr
```

```
# Choose ONE model

MODEL_NAME="HuggingFaceTB/SmolLM2-360M-Instruct"

# MODEL_NAME="distilgpt2"

# MODEL_NAME="Qwen/Qwen2.5-0.5B-Instruct"

tokenizer=AutoTokenizer.from_pretrained(MODEL_NAME)
model=AutoModelForCausalLM.from_pretrained(MODEL_NAME)

print("Loaded:",MODEL_NAME)
```

```
def predict_next_words(prompt,top_k=5):
    inputs=tokenizer(prompt,return_tensors="pt")
    with torch.no_grad():
        outputs=model(**inputs)
    logits=outputs.logits[:,-1,:]
    probs=torch.softmax(logits,dim=-1)
    values,indices=torch.topk(probs,top_k)
    results=[]
    for s,i in zip(values[0],indices[0]):
        results.append((tokenizer.decode([i]).strip(),float(s)))
    return results
```

```
prompt="Artificial Intelligence will"
for i,(w,s) in enumerate(predict_next_words(prompt),1):
    print(f"{i}. {w} ({s:.4f})")
```

```
def predict(prompt):
    out=""
    for i,(w,s) in enumerate(predict_next_words(prompt),1):
        out+=f"{i}. {w} ({s:.4f})\n"
    return out

demo=gr.Interface(
fn=predict,
inputs=gr.Textbox(lines=2,label="Prompt"),
outputs=gr.Textbox(label="Top 5 Next Word Predictions"),
title="🚀 Samsung GenAI Lab",
description="Transformer Next Token Prediction",
examples=[
["Artificial Intelligence will"],
["Samsung is developing"],
["Cloud computing enables"],
["Transformers work by"],
["Data Science is"]
]
)
demo.launch(share=True)
```
